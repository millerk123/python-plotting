```python
# First, look at EPW -->  The horizontal axes is the x-position, and the vertical axes is time.  
# In the beginning of the simulation, the plasma wave is going strictly from left to right  
```

```python
# read the history file, plot it vs time and space
test4=read_hdf('/Users/franktsung/Documents/work/summer_school_2017/examples/ldi/ldi-r10-epw.h5')
test4.data=test4.data
plt.figure()
plotme(test4)
# plt.figure()
# plt.show()
```
